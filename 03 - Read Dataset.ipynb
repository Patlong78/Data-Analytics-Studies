```python
import pandas as pd
df = pd.read_excel("Global_Power_Emissions_Database_v1.0.xlsx", sheet_name="GPED_v1.0_Plant Level")
df.head()
```

```python
df.columns = df.iloc[0]
df = df[1:].reset_index(drop=True)
df.head()
```

```python
df.info()
```

```python
# the first row should be the header, but it is not. We need to set the first row as the header and then remove it from the data. We can do this by using the following code:
df.columns = df.iloc[0]
```

```python
df.head()
```

```python
# Delete row 0
df = df.drop(0).reset_index(drop=True)
```

```python
df.head()
```

```python
col="Plant Name"
print(f"Number of unique values: {df[col].nunique()}")
print(f"Unique values and their counts:\n{df[col].value_counts()}\n")
```

```python
col="Fuel Types"
print(f"Number of unique values: {df[col].nunique()}")
print(f"Unique values and their counts:\n{df[col].value_counts()}\n")
```

```python
col="Country (or Region)"
print(f"Number of unique values: {df[col].nunique()}")
print(f"Unique values and their counts:\n{df[col].value_counts()}\n")  
```

```python
# Delete row Plant Name column
df = df.drop(columns=["Plant Name"])
```

```python
df.head()
```

```python
# convert Fuel Types column to an indices 0 to 4
df["Fuel Types"] = df["Fuel Types"].astype("category").cat.codes
```

```python
df.info()
```

```python
# Convert Country (or Region) into an integer in alphabetical order
df["Country (or Region)"] = df["Country (or Region)"].astype("category").cat.codes
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
